#### Task4: Sorting an array in ascending order
If user supplies an array(list) like arr = [5, 2, 8, 7, 1]. 


The output should be like:

Elements of original array: 
5 2 8 7 1 

Elements of array sorted in ascending order: 
1 2 5 7 8 

Note: We need to write our own sorting algorithm using sort function of list is not acceptable.

```python
#Your code goes here
NumList = []

Number = int(input("Please enter the Total Number of List Elements: "))
for i in range(Number):
    value = int(input("Please enter the Value of %d Element : " %i))
    NumList.append(value)

for i in range (Number):
    for j in range(i + 1, Number):
        if(NumList[i] > NumList[j]):
            temp = NumList[i]
            NumList[i] = NumList[j]
            NumList[j] = temp

print("Element After Sorting List in Ascending Order is : ", NumList)
```

#### Task5: Sorting an array in descending order

If user supplies an array(list) like arr = [5, 2, 8, 7, 1]. 


The output should be like:

Elements of original array: 
5 2 8 7 1 

Elements of array sorted in descending order: 
8 7 5 2 1 

Note: We need to write our own sorting algorithm. Using sort function of list is not acceptable.


```python
#Your code goes here 
arr=[]
N=int(input("enter total number of elements: "))
for i in range(N):
    num=input("enter num: ")
    arr.append(num)
for i in range(N):
    for j in range(i+1,N):
        if arr[i]<arr[j]:
            temp=arr[j]
            arr[j]=arr[i]
            arr[i]=temp
print(f"Elements of array sorted in descending order {arr} ")
```
